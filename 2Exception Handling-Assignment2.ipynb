```python
# Ans - The arithmetic error occurs when an error is encountered during numeric calculations in Python.
# This includes Zerodivision Error and Floating point error. 
# Zero division error is raised when you divide a numeric value by zero.
```

```python
# ZeroDivision Error
try:
    a = 23
    a/0
except ZeroDivisionError as e:
    print(e)
```

```python
# floating point error - It happens because it is the way the underlying c platform handles floating-point numbers 
# and ultimately with the inaccuracy, we'll always have been writing down numbers as a string of fixed number of digits.
```

```python
try:
    a = 1.0
    b = 1.2
    c = a+b
    print(c)
except FloatingPointError as e:
    print(e)
```

```python
# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
```

```python
# Ans - Lookup Error class forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively.
```

```python
# An IndexError is raised when a sequence reference is out of range.
```

```python
try: 
    list = [3, 6, 7, 9, "sudh"]
    print(list[7])
except IndexError as e:
    print(e)
```

```python
# A KeyError is raised when a value is not found as a key of a dictionary. 
# It denotes that there was an issue retrieving the key you were looking for. 
```

```python
try :
    d1 = {'name' : ['Krish', 'Sudh', 'Narayan'], 'Universities' :('Mumbai', 'US', 'UK', 'Germany')}
    print(d1['name'])
    
except KeyError as e:
    print(" Key error has occured")
```

```python
# Q6. List down some best practices for exception handling in python.
```

```python
# Ans - Some of the Best practices are:
# 1. Use always a specific exception. Writing a superclass exception is not a good practice
# 2. Print always a valid / meaningful message. This helps the programmer during debugging
# 3. Keep logging. This helps in generating all possible information during execution/running of the program and that helps during debugging.
# 4. Always avoid writing multiple exception handling statements
# 5. Prepare proper documentation for the given project/program
# 6. Clean up all resources eg. always close all open files even if the files are open in try block 
```
